```python
#Calculate Column Line a Stock Report 


#7 Bandung Stock
stock['stock_value_BDG'] = stock.cost_price_new*stock.real_stock_bandung
stock['(Day)_Average_Sellout_BDG'] = (stock[[('month 1', 'Bandung'),\
                                             ('month 2', 'Bandung'),('month 3', 'Bandung'),]].sum(axis=1))/days
stock['(Month)_Average_Sellout_BDG'] = stock['(Day)_Average_Sellout_BDG']*30
stock['TurnOver_Days_BDG'] = stock['real_stock_bandung']/stock['(Day)_Average_Sellout_BDG']


#Stock Days BDG

stock.loc[((stock['TurnOver_Days_BDG'])>360) ,'Stock_Days_BDG'] ='>360 Days'
stock.loc[((stock['TurnOver_Days_BDG']<=360) & (stock['TurnOver_Days_BDG']>180)) ,'Stock_Days_BDG'] ='180-360 Days'
stock.loc[((stock['TurnOver_Days_BDG']<=180) & (stock['TurnOver_Days_BDG']>90)) ,'Stock_Days_BDG'] ='90-180 Days'
stock.loc[((stock['TurnOver_Days_BDG']<=90) & (stock['TurnOver_Days_BDG']>60)) ,'Stock_Days_BDG'] ='60-90 Days'
stock.loc[((stock['TurnOver_Days_BDG']<=60) & (stock['TurnOver_Days_BDG']>30)) ,'Stock_Days_BDG'] ='30-60 Days'
stock.loc[((stock['TurnOver_Days_BDG']<=30) & (stock['TurnOver_Days_BDG']>14)) ,'Stock_Days_BDG'] ='15-30 Days' 
stock.loc[((stock['TurnOver_Days_BDG']<15) & (stock['TurnOver_Days_BDG']>0)) ,'Stock_Days_BDG'] ='<14 Days'
stock.loc[((stock['TurnOver_Days_BDG']<=0)) ,'Stock_Days_BDG'] ='0 Days'



#OOS Status BDG
stock.loc[((stock['real_stock_bandung'])>(stock['(Month)_Average_Sellout_BDG'])),'OOS_Status_BDG'] ='Over Stock'
stock.loc[((stock['TurnOver_Days_BDG']<=22) & (stock['TurnOver_Days_BDG']>7)) ,'OOS_Status_BDG'] ='Need Stock'
stock.loc[((stock['TurnOver_Days_BDG']<=7) & (stock['TurnOver_Days_BDG']>4)) ,'OOS_Status_BDG'] ='Low Stock'
stock.loc[((stock['TurnOver_Days_BDG']<=4) & (stock['TurnOver_Days_BDG']>0)) ,'OOS_Status_BDG'] ='Very Low Stock'
stock.loc[((stock['TurnOver_Days_BDG']<=0)) ,'OOS_Status_BDG'] ='OOS'
stock.loc[((stock[('month 1', 'Bandung')]==0) & (stock[('month 2', 'Bandung')]==0) & (stock[('month 3', 'Bandung')]==0) &\
           (stock['(Month)_Average_Sellout_BDG']==0) & (stock['(Day)_Average_Sellout_BDG']==0)) ,'OOS_Status_BDG']\
            ='No Stock No Sale'

stock['Stock_Days_BDG'] = stock['Stock_Days_BDG'].fillna('NA')
stock['TurnOver_Days_BDG'] = stock['TurnOver_Days_BDG'].fillna('NA')
stock.replace(np.inf, 'NA', inplace=True)



stock.loc[((stock['real_stock_bandung']>0) & (stock['TurnOver_Days_BDG']=='NA')) ,'OOS_Status_BDG'] ='Never Selling'
stock['OOS_Status_BDG'] = stock['OOS_Status_BDG'].fillna('OK')
```
